```python
columns = dataset.columns
```

```python
dataset.describe()
```

```python
# Number of columns and rows in the subplot matrix
n_columns = 5
n_lines = (len(columns) + n_columns - 1) // n_columns

# create matrix graphics
fig, axis = plt.subplots(n_lines, n_columns, figsize=(23, 25))

# adjusts the spaces between the subplots
plt.subplots_adjust(hspace=0.5, wspace=0.4)

for i, col in enumerate(columns):
    l = i // n_columns
    c = i % n_columns
    ax = axis[l, c]
    ax.hist(dataset[col], bins=8, linewidth=0.5, edgecolor="brown")
    ax.set_title(f'variable {col}')

# removes empty subplots
for i in range(len(columns), n_lines * n_columns):
    fig.delaxes(axis.flatten()[i])

plt.savefig('variables.png')

plt.show()
```
